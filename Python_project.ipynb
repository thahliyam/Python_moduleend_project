```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_excel("excel python.xlsx")
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
df = pd.DataFrame(data)
df['Height'] = np.random.randint(150, 181, size=len(data))

print(data)

```

```python
# Replace NaN values in college,salary columns with zero values
df[['College', 'Salary']]= df[['College', 'Salary']].fillna(0)  

print(df)
```

```python
# Distribution of employees across each team
team_distribution = df['Team'].value_counts()
team_percentage = (team_distribution / df.shape[0]) * 100

# Displaying the distribution and percentage
distribution_df = pd.DataFrame({'Employee Count': team_distribution, 'Percentage (%)': team_percentage.round(2)})
print(distribution_df)

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=distribution_df.index, y=distribution_df['Employee Count'], palette='viridis')
plt.xticks(rotation=45)
plt.title('Employee Distribution Across Teams')
plt.ylabel('Number of Employees')
plt.xlabel('Teams')
plt.show()
```

```python
# Segregating employees based on their positions
position_distribution = df['Position'].value_counts()

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=position_distribution.values, y=position_distribution.index, 
            hue=position_distribution.index, palette='coolwarm', legend=False)  # Added hue and disabled legend
plt.title('Employee Distribution Based on Positions')
plt.xlabel('Number of Employees')
plt.ylabel('Positions')
plt.show()
```

```python
# Creating age bins
age_bins = [20, 25, 30, 35, 40, 45, 50, 60]
age_labels = ['20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50+']
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

# Finding the predominant age group
age_group_distribution = df['Age Group'].value_counts().sort_index()

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=age_group_distribution.index, y=age_group_distribution.values, 
            hue=age_group_distribution.index, palette='Set2', legend=False)
plt.title('Predominant Age Group Among Employees')
plt.xlabel('Age Groups')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Salary expenditure by Team
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)

# Salary expenditure by Position
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)

# Displaying the top team and position
print("Team with Highest Salary Expenditure:")
print(team_salary.head(1))

print("\nPosition with Highest Salary Expenditure:")
print(position_salary.head(1))

# Visualization 
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Team Salary Expenditure
sns.barplot(y=team_salary.index, x=team_salary.values, 
            hue=team_salary.index, palette='magma', ax=axes[0], legend=False)  # Added hue and disabled legend
axes[0].set_title('Salary Expenditure by Team')
axes[0].set_xlabel('Total Salary')
axes[0].set_ylabel('Teams')

# Position Salary Expenditure
sns.barplot(y=position_salary.index, x=position_salary.values, 
            hue=position_salary.index, palette='cividis', ax=axes[1], legend=False)  # Added hue and disabled legend
axes[1].set_title('Salary Expenditure by Position')
axes[1].set_xlabel('Total Salary')
axes[1].set_ylabel('Positions')

plt.tight_layout()
plt.show()
```

```python
# Checking correlation between Age and Salary
correlation = df[['Age', 'Salary']].corr().iloc[0, 1]
print(f"Correlation between Age and Salary: {correlation:.2f}")

# Visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df, hue='Team', palette='Spectral', alpha=1)
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
